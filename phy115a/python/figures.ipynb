```python
from math import pi

z0 = 10

z = np.linspace(0,10*pi,1000)
mask = (z<z0)&(z>0)
za = z[mask]
ya = sqrt((z0/za)**2-1)

plt.plot(za,ya,"r-",label="$\sqrt{(z_0/z)^2 - 1}$")

mask = (z>0)&(z<pi/2)
zb1 = z[mask]
yb1 = tan(zb1)
plt.plot(zb1,yb1,"b--", label="tan$(z)$")

mask = (z>pi)&(z<3*pi/2)
zb2 = z[mask]
yb2 = tan(zb2)
plt.plot(zb2,yb2,"b--")

mask = (z>2*pi)&(z<5*pi/2)
zb3 = z[mask]
yb3 = tan(zb3)
plt.plot(zb3,yb3,"b--")

mask = (z>3*pi)&(z<7*pi/2)
zb4 = z[mask]
yb4 = tan(zb4)
plt.plot(zb4,yb4,"b--")

plt.ylim(0,10)

plt.xlabel("z")
plt.ylabel("f(z)")
plt.legend(loc=1, frameon=False)

plt.savefig("finitewell.pdf")
```

```python
a=1 
v=-3 # in units of E*
#v=10 # in units of E*
#v=0.5 # in units of E*
n = int(np.floor(np.sqrt(np.abs(v)))+1)
print("minimum n = ", n)
n=2.0
print("n         = ", n)

eta=0
k=0



eta = np.sqrt(1+v/n**2)
print("eta =       ", eta)
k = n * pi / (2*a*eta)
print("k =         ", k)
print("eta*k =     ", eta*k)
```

```python
# global variables:
eta=0
k=0
z=0
cos_theta=0
sin_theta=0
T=0
R=0
phi=0

def update_globals():
    global eta,k,z,cos_theta,sin_theta,T,R,phi
    eta = np.sqrt(1+v/n**2)
    k = n * pi / (2*a*eta)
    z = (eta**2-1)/(2*eta)
    cos_theta=np.cos(2*eta*k*a)
    sin_theta=np.sin(2*eta*k*a)
    tmp = (z*sin_theta)**2
    T = 1/(1+tmp)
    R = tmp*T
    phi = np.arctan2(sqrt(z**2+1)*sin_theta,cos_theta)
    
    
def dump_globals():
    print("n         = ", n)
    print("eta       = ", eta)
    print("k         = ", k)
    print("eta*k     = ", eta*k)
    print("z         = ", np.round(z,5)+0)
    print("cos_theta = ", cos_theta)
    print("sin_theta = ", np.round(sin_theta,5)+0)
    print("R   =", R)
    print("T   =", T)
    print("R+T =",R+T)
    print("phi=",phi)  
    

    
update_globals()
dump_globals()
```

```python
#resonant version
A=1
def psi_real(x):
    S = (-1)**n
    if (x<-a):
        return A*np.cos(k*(x+a))
    if (x>a):
        return S*A*np.cos(k*(x-a))    
    else:
        return A*np.cos(eta*k*(x+a))
    
def psi_imag(x):
    S = (-1)**n
    if (x<-a):
        return A*np.sin(k*(x+a))
    if (x>a):
        return S*A*np.sin(k*(x-a))   
    else:
        return A*np.sin(eta*k*(x+a))/eta
```

```python
#non-resonant version
A = 1
z = (eta**2-1)/(2*eta)
print("z = ", z)
cos_theta=np.cos(2*eta*k*a)
sin_theta=np.sin(2*eta*k*a)
print("cos_theta = ", cos_theta)
print("sin_theta = ", sin_theta)
tmp = (z*sin_theta)**2
T = 1/(1+tmp)
R = tmp*T
print("R   =", R)
print("T   =", T)
print("R+T =",R+T)
phi = np.arctan2(sqrt(z**2+1)*sin_theta,cos_theta)
print("phi=",phi)  
      
def psi_real(x):
    if (x<-a):
        return A*np.cos(k*(x+a))+A*sqrt(R)*sin(-k*(x+a)+phi)
    if (x>a):
        return A*sqrt(T)*cos(k*(x-a)+phi)
    else:
        return A*sqrt(T)*(cos(phi)*cos(eta*k*(x-a)) - sin(phi)*sin(eta*k*(x-a))/eta)    
    
def psi_imag(x):
    if (x<-a):
        return A*np.sin(k*(x+a))-A*sqrt(R)*cos(-k*(x+a)+phi)
    if (x>a):
        return A*sqrt(T)*sin(k*(x-a)+phi)
    else:
        return A*sqrt(T)*(sin(phi)*cos(eta*k*(x-a)) + cos(phi)*sin(eta*k*(x-a))/eta)
    
def ref_real(x):
    if (x<-a):
        return A*sqrt(R)*sin(-k*(x+a)+phi)
    if (x>a):
        return 0
    else:
        return 0

def ref_imag(x):
    if (x<-a):
        return -A*sqrt(R)*cos(-k*(x+a)+phi)
    if (x>a):
        return 0
    else:
        return 0


    
    
def ref_real_delta(x,delta):
    return ref_real(x)*cos(delta)-ref_imag(x)*sin(delta)

    
    
```

```python
def psi_real_delta(x,delta):
    return psi_real(x)*cos(delta)-psi_imag(x)*sin(delta)
    
def psi_imag_delta(x,delta):
    return psi_imag(x)*cos(delta)+psi_real(x)*sin(delta)
    
def psi_magn_delta(x,delta):
    r = psi_real_delta(x,delta)
    i = psi_imag_delta(x,delta)
    return sqrt(r*r+i*i)
```

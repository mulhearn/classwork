```python
from scipy import stats
MU     = 1.0
MIN    = 0
MAX    = 11
NBINS  = 11
NTHROW = 1000
s = np.random.poisson(MU, NTHROW)
mu = np.mean(s)
counts,bins = np.histogram(s,range=(MIN,MAX),bins=NBINS)
cbins = bins[:-1]
err = counts**0.5
plt.errorbar(cbins,counts,fmt="ko",yerr=err,label="Simulated Data")
xp=np.linspace(MIN,MAX,MAX-MIN+1)
yp=NTHROW*stats.poisson.pmf(xp, mu)

xg=np.linspace(MIN,MAX,100)
yg=NTHROW*stats.norm.pdf(xg, loc=mu, scale=mu**0.5)

plt.plot(xp,yp,"bo-",label="Poisson")
plt.plot(xg,yg,"b--",label="Gaussian")
plt.title("Monte Carlo Study for $\lambda = 1$")
plt.legend()
plt.xlabel("Counts")
plt.ylabel("Occurances")
plt.savefig("montecarlo_eg.pdf")


```

```python
T = 3.0 # mseconds

def f(x):
   return 20*sin(6*2*pi*x/T) + 10*sin(9*2*pi*x/T)
max_samples = 20
fs = max_samples/T
x = T*np.arange(max_samples)/max_samples
xfin = T*np.arange(0,max_samples,0.1)/max_samples
y = f(x)
yfin = f(xfin)
plot(x,y,"o")
plot(xfin,yfin,"--")
plt.ylabel('amplitude [mV]')
plt.xlabel('time [ms]')
plt.savefig("periodogram_ts.png")
```

```python
from scipy import signal
#fs=76.900 # sample frequency in Hz
f, psd = signal.periodogram(y, fs)
#plt.semilogy(f, psd)
plt.ylim(0,700)
plt.plot(f, psd,"o")
plt.xlabel('frequency [kHz]')
plt.ylabel('PSD [mV**2/kHz]')
plt.savefig("periodogram_eg.png")
```

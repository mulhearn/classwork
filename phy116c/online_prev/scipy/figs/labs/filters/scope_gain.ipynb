```python

f = 5E3 # 5 kHz
phi = 0.25
w = 2*pi* f *(1E-6) # in units of 1/ms
t = np.arange(0,200,1, dtype=np.float)

tx = [100]
ty = [0]

ya = -2.0*sin(w*t)
yb = -(2.0/sqrt(2.0))*sin(w*t-pi/4.0)
plt.plot(t,ya,"--",label="Vin",color="red")
plt.plot(t,yb,"-",label="Vout",color="blue")
plt.plot(tx,ty,"o",label="Trigger")
plt.xlabel("Time [ms]")
plt.ylabel("Voltage [V]")
plt.legend(numpoints=1,loc=4)
plt.axhline(0,color="black")
plt.axvline(100,color="green")
plt.axvline(125,color="green")
plt.axhline(1.414,color="green")
plt.axhline(2.0,color="green")
plt.ylim(-2.5,2.5)
plt.xticks(np.arange(0, 201, 25.0))
plt.yticks(np.arange(-2.5, 2.51, 0.5))
plt.savefig("scope_gain.pdf")


 
```

```python
f = 5E3 # 5 kHz
phi = 0.25
w = 2*pi* f *(1E-6) # in units of 1/ms
t = np.arange(0,200,1, dtype=np.float)

ya = 1.0*sin(w*t)
yb = 1.0*sin(w*t-pi/2.0)
yc = 1.0*sin(w*t-pi/4.0)
yd = 1.0*sin(w*t)
plt.xlim(-2.,2.0)
plt.ylim(-1.5,1.5)
plt.plot(ya,yb,":",label=r'$\phi=\pi/2$',color="black")
plt.plot(ya,yc,"--",label="$\phi=\pi/4$",color="green")
plt.plot(ya,yd,"-",label="$\phi=0$",color="blue")
#plt.rc('text', usetex=False)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(loc=2)
plt.savefig("scope_xy.pdf")
```

```python
ff = np.arange(-2,2.,0.01)
ff = 10**ff
ya = 1.0/np.sqrt(1+ff*ff)
yb = 1.0/np.sqrt(1+1.0/(ff*ff))

#f=np.array([1.0/100,1.0/30,1.0/10.0,1/3.0,1.0,3.0,10.0,30.0,100.0])
#y  = -10.0*np.log10(1+f*f)

plt.xscale("log")
plt.plot(ff,ya,"-",label="lowpass", color="black")
plt.plot(ff,yb,"--",label="highpass",color="black")
plt.xlabel("f/f0 [1]")
plt.ylabel("Vout/Vin [1]")
plt.legend(loc=8)
```

```python
ff = np.arange(-2,2.,0.01)
ff = 10**ff
ya = -10.0*np.log10(1+ff*ff)
yb = -10.0*np.log10(1+1.0/(ff*ff))

plt.xscale("log")
plt.plot(ff,ya,"-",label="lowpass", color="black")
plt.plot(ff,yb,"--",label="highpass",color="black")
plt.xlabel("f/f0")
plt.ylabel("Vout/Vin [dB]")
plt.legend(loc=8)

plt.savefig("bode.pdf")
```

```python
ff = np.arange(-2,2.,0.01)
ff = 10**ff
ya = -np.arctan(ff)
yb = np.arctan(1/ff)


fig = plt.figure()
ax  = fig.add_subplot(111)


unit   = 0.25
y_tick = np.arange(-0.5, 0.5+unit, unit)
y_label = [r"$-\pi/2$", r"$-\pi/4$", "0", r"$\pi/4$",   r"$\pi/2$"]
ax.set_yticks(y_tick*np.pi)
ax.set_yticklabels(y_label)

ax.set_ylim([-pi/2.0-0.1, pi/2.0+0.1])

ax.semilogx()
ax.plot(ff,ya,"-",label="lowpass", color="black")
ax.plot(ff,yb,"--",label="highpass",color="black")
ax.set_xlabel("f/f0")
ax.set_ylabel("$\mathbf{\phi}$")
ax.legend(loc=1)
fig.savefig("phase.pdf")


```

```python
np.random.seed(2019)
```

```python
NEVT = 200
dat = np.random.normal(loc=10.0,scale=3.0,size=NEVT)
print(dat[:5])
```

```python
counts,edges = np.histogram(dat,bins=10,range=(0,20))
cbins = (edges[1:] + edges[:-1])/2.0
print("counts:     ", counts)
print("total:      ", np.sum(counts))
print("bin edges:  ", edges)
print("cbins:      ", cbins)
```

```python
plt.plot(cbins,counts,"ko")
plt.xlim(0,20)
plt.xlabel("Outcome")
plt.ylabel("Occurances")
```

```python
errs = counts**0.5
plt.errorbar(cbins,counts,yerr=errs,fmt="ko")
plt.xlim(0,20)
plt.xlabel("Outcome")
plt.ylabel("Occurances")
```

```python
from scipy import stats
BINW = 2.0
plt.errorbar(cbins,counts,yerr=errs,fmt="ko",
             label="Simulated Data")
plt.xlabel("Outcome")
plt.ylabel("Occurances")
xpred = np.linspace(0,20,100)
mean = np.mean(dat)
sigma = np.var(dat)**0.5
ypred = BINW * NEVT * stats.norm.pdf(xpred,loc=mean,scale=sigma) 
plt.plot(xpred, ypred,"b--",label="Gaussian")
plt.legend()
plt.savefig("gaussian_eg.pdf")
```

```python
NEVT = 10000
BINS = 50
BINW = 20.0/BINS
dat = np.random.normal(loc=10.0,scale=3.0,size=NEVT)
counts,edges = np.histogram(dat,bins=BINS,range=(0,20))
cbins = (edges[1:] + edges[:-1])/2.0
errs = counts**0.5
plt.errorbar(cbins,counts,yerr=errs,fmt="ko",
             label="Simulated Data")
plt.xlabel("Outcome")
plt.ylabel("Occurances")
xpred = np.linspace(0,20,100)
mean = np.mean(dat)
sigma = np.var(dat)**0.5
ypred = BINW * NEVT * stats.norm.pdf(xpred,loc=mean,scale=sigma) 
plt.plot(xpred, ypred,"b--",label="Gaussian")
plt.legend()
```

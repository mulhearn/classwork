```python
%matplotlib inline

from math import sin
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

plt.xlim(-5, 5)
plt.ylim(-5, 5)
ax.set_xticks([-5,-4,-3,-2,-1,1,2,3,4,5])
ax.set_yticks([-5,-4,-3,-2,-1,1,2,3,4,5])
ax.grid(True)

A = 4
k = 2 * np.pi / 4.0
w = 2 * np.pi / 2.0

x = np.linspace(-5, 5, 500)
plt.plot(x, A*np.sin(k*x), color="blue")
plt.savefig("sinx")
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

plt.xlim(0, 4)
plt.ylim(0, 13)
ax.set_xticks([1,2,3,4])
ax.set_yticks([2,5,8,11])
ax.set_xticklabels(["L/4","L/2","3L/4","L"])
ax.set_yticklabels(["$\lambda=L/2$","$\lambda = 2L/3$","$\lambda= L$","$\lambda=2L$"])
ax.grid(True)

A = 1
k = np.pi / 4.0

x = np.linspace(-5, 5, 500)
plt.plot(x, 11 + np.sin(k*x), color="red")
plt.plot(x, 11 + 0*x, color="green")
plt.plot(x, 11 - np.sin(k*x), color="blue")

plt.plot(x, 8 + np.sin(2*k*x), color="red")
plt.plot(x, 8 + 0*x, color="green")
plt.plot(x, 8 - np.sin(2*k*x), color="blue")

plt.plot(x, 5 + np.sin(3*k*x), color="red")
plt.plot(x, 5 + 0*x, color="green")
plt.plot(x, 5 - np.sin(3*k*x), color="blue")

plt.plot(x, 2 + np.sin(4*k*x), color="red")
plt.plot(x, 2 + 0*x, color="green")
plt.plot(x, 2 - np.sin(4*k*x), color="blue")



plt.savefig("harmonics.png")

```

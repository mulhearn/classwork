```python
P = 2.0
x = linspace(-3, 3, 2000)

def d(x, a):
    if (x==0):
        return a/3.1415
    return sin(a*x)/x/3.1415

ya = [d(xi, 2) for xi in x]
yb = [d(xi, 5) for xi in x]
yc = [d(xi, 10) for xi in x]

xlabel("x")
ylabel("y")
plot(x, ya, "g", label="a=2")
plot(x, yb, "r", label="a=5")
plot(x, yc, "b", label="a=10")
legend()

savefig("sinc.png")

```

```python
def fourier_series(x, fx, n, P):
    stot = 0.0
    ctot = 0.0
    for i in range(x.size-1):
        stot += (fx[i] * sin(2*pi*n*x[i]/P) + fx[i+1] * sin(2*pi*n*x[i+1]/P)) * (x[i+1] - x[i])
        ctot += (fx[i] * cos(2*pi*n*x[i]/P) + fx[i+1] * cos(2*pi*n*x[i+1]/P)) * (x[i+1] - x[i])
    norm = x[x.size-1] - x[0];
    return stot / norm, ctot / norm
```

```python
s1, c1 = fourier_series(x, y, 1, P)
s2, c2 = fourier_series(x, y, 2, P)
s3, c3 = fourier_series(x, y, 3, P)
s4, c4 = fourier_series(x, y, 4, P)
s5, c5 = fourier_series(x, y, 5, P)
s6, c6 = fourier_series(x, y, 6, P)
s7, c7 = fourier_series(x, y, 7, P)
s9, c9 = fourier_series(x, y, 9, P)
s11, c11 = fourier_series(x, y, 11, P)
s13, c13 = fourier_series(x, y, 12, P)
s15, c15 = fourier_series(x, y, 12, P)
s17, c17 = fourier_series(x, y, 12, P)
s19, c19 = fourier_series(x, y, 12, P)


#note that the cosine terms in this case are all zero:
print "cos terms:  ", c1, c2, c3, c4, c5, c6, c7
print "sin terms:  ", s1, s2, s3, s4, s5, s6, s7
```

```python
from scipy.stats import norm
NSUM    = 100000
NBIN    = 30
NSIG    = 5.0

def throw_flat(nthrow):
    x = np.array([1,2,3])
    x = np.random.uniform(-1.,1.0,size=(NSUM,nthrow))
    sums = np.mean(x,axis=1)
    sums = sums / np.var(sums)**0.5
    h,bins = np.histogram(sums,bins=NBIN,range=(-NSIG,NSIG),density=True)
    cbins = (bins[:-1] + bins[1:])/2.0
    return h,cbins

ha,ca = throw_flat(1)
hb,cb = throw_flat(2)
hc,cc = throw_flat(3)
hd,cd = throw_flat(10)

xf = np.linspace(-5.0,5.0,100)
yf = norm.pdf(xf)

ax = plt.subplot(1, 3, 1)
ax.set_title("N=1")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--",label="G"),plt.axis('equal')
ax.plot(ca,ha,"r.",label="A")
ax.legend()
ax = plt.subplot(1, 3, 2)
ax.set_title("N=2")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--")
ax.plot(cb,hb,"r.")
ax = plt.subplot(1, 3, 3)
ax.set_title("N=3")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--")
ax.plot(cc,hc,"r.")

```

```python
#
# Monte Carlo Data Generation Example:  Binomial Process
#
EFF    = 0.5   # efficiency of each trial
NTRIAL = 5     # the number of trials for each event
NTHROW = 15    # the number of simulated outcome to generate

# MC event function:  count the number of successes after NTRIAL attempts with efficiency EFF.
def mc():
    return np.sum(np.random.uniform(0.0,1.0,NTRIAL)<EFF)

# produce pseudo-data by calling mc() repeatedly to produce NTHROW outcomes"
successes = np.array([mc() for i in range(NTHROW)])

print("simulated outcomes [number of successes]:  ", successes)


#
# Alternate 2-D array version:
#
x = np.random.uniform(A,B,size=(NTHROW,NTRIAL))
successes = np.sum(x>EFF,axis=1)
print("simulated outcomes [number of successes]:  ", successes)


```

```python
# HW Distributions:  A simulated Binomial process

EFF    = 0.75   # efficiency of each trial
NTRIAL = 5     # the number of trials for each event
NTHROW = 1000    # the number of simulated outcome to generate

# MC event function:  count the number of successes after NTRIAL attempts with efficiency EFF.
def mc():
    return np.sum(np.random.uniform(0.0,1.0,NTRIAL)<EFF)

# produce pseudo-data by calling mc() repeatedly to produce NTHROW outcomes"
successes = np.array([mc() for i in range(NTHROW)])
h,bins = np.histogram(successes,bins=NTRIAL+1,range=(0,NTRIAL+1))
bins = bins[:-1]
plt.plot(bins,h,"o")
print(bins)

y = binom.pmf(bins,NTRIAL, EFF) * NTHROW
plt.plot(bins,y,"-")

```

```python
# HW DIstributions:  a simulated Poisson process

from scipy.stats import poisson
LAMBDA = 2.5   # the mean number of events expected
NTRIAL = 1000  # when simulated as a Binomial process, the number of trials.
               # Poisson process has NTRIAL -> infinity  
NTHROW = 10000  # the number of simulated events to generate
UPPER  = 10    # Upper limit for histogram


# calculate the efficiency of the binomial process that gives mean LAMBDA:
EFF = float(LAMBDA) / NTRIAL

# MC event function:  count the number of successes after NTRIAL attempts with efficiency EFF.
def mc():
    return np.sum(np.random.uniform(0.0,1.0,NTRIAL)<EFF)

# produce pseudo-data by calling mc() repeatedly to produce NTHROW outcomes"
counts = np.array([mc() for i in range(NTHROW)])

h,bins = np.histogram(counts,bins=UPPER+1,range=(0,UPPER+1))
bins = bins[:-1]
plt.plot(bins,h,"o")

y = poisson.pmf(bins,LAMBDA)* NTHROW
plt.plot(bins,y,"-")

```

```python
# HW Distributions:  a simulated Gaussian process

from scipy.stats import norm
MEAN   = 10.0     # the simulated mean
SIGMA  = 2.0      # the simulated sigma
NTRIAL = 100000     # the number of simulated events to generate
UPPER  = 20
BINS   = 10
#BINS = 50
#BINS = 11 #(reveals bias from comparing central value to integral of bin)

BIN_WIDTH = float(UPPER) / BINS
data = np.random.normal(MEAN,SIGMA,size=NTRIAL)

h,bins = np.histogram(data,bins=BINS,range=(0,UPPER))
cbins = (bins[1:] + bins[:-1]) / 2.0
plt.plot(cbins,h,"o")


fbins = np.linspace(0,UPPER,400)
y = norm.pdf(fbins,MEAN,SIGMA) * NTRIAL * BIN_WIDTH
plt.plot(fbins,y,"-")

```

```python
# demonstration of central limit theorem

from scipy.stats import norm
NTHROW  = 20
NSUM    = 100000
NBIN    = 31
A = 0.0
B = 10.0


x = np.random.uniform(A,B,size=(NSUM,NTHROW))
sums = np.mean(x,axis=1)

sig  = np.var(sums)**0.5
mean = np.mean(sums)
rng = (mean-5*sig, mean+5*sig)
h,bins = np.histogram(sums,bins=NBIN,range=rng)
cbins = (bins[:-1] + bins[1:])/2.0
sbin = bins[1]-bins[0]
plt.plot(cbins,h,"o")

fbins = np.linspace(rng[0],rng[1],200)
gfit = sbin*NSUM*norm.pdf((fbins-mean)/sig)/sig
#plt.plot(fbins,gfit)

pmean = (B + A) / 2.0
psig  = (B - A) / 2 / ((3.0 * NTHROW))**0.5
print("mean pred:   ",  pmean, " measured:  ", mean)
print("sigma pred:  ", psig,   " measured:  ", sig)
pred = sbin*NSUM*norm.pdf((fbins-pmean)/psig)/psig
plt.plot(fbins,pred)

#plt.semilogy()

```

```python
from scipy.stats import norm
NSUM    = 100000
NBIN    = 30
NSIG    = 5.0

def throw_flat(nthrow):
    x = np.array([1,2,3])
    x = np.random.uniform(-1.,1.0,size=(NSUM,nthrow))
    sums = np.mean(x,axis=1)
    sums = sums / np.var(sums)**0.5
    h,bins = np.histogram(sums,bins=NBIN,range=(-NSIG,NSIG),density=True)
    cbins = (bins[:-1] + bins[1:])/2.0
    return h,cbins

ha,ca = throw_flat(1)
hb,cb = throw_flat(2)
hc,cc = throw_flat(3)
hd,cd = throw_flat(10)

xf = np.linspace(-5.0,5.0,100)
yf = norm.pdf(xf)

plt.subplots_adjust(hspace=0.5)
ax = plt.subplot(2, 2, 1)
ax.set_title("N=1")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--",label="G")
ax.plot(ca,ha,"r.",label="A")
ax.legend()
ax = plt.subplot(2, 2, 2)
ax.set_title("N=2")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--")
ax.plot(cb,hb,"r.")
ax = plt.subplot(2, 2, 3)
ax.set_title("N=3")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--")
ax.plot(cc,hc,"r.")
ax = plt.subplot(2, 2, 4)
ax.set_title("N=10")
ax.set_ylim(0.,0.45)
ax.plot(xf,yf,"--")
ax.plot(cd,hd,"r.")







```

```python
from scipy.stats import binom
x = np.arange(0,11)

ya = binom.pmf(x,1,0.5)
yb = binom.pmf(x,3,0.5)
yc = binom.pmf(x,5,0.5)
yd = binom.pmf(x,10,0.5)

xa = x[ya > 0]
ya = ya[ya > 0]
xb = x[yb > 0]
yb = yb[yb > 0]
xc = x[yc > 0]
yc = yc[yc > 0]
xd = x[yd > 0]
yd = yd[yd > 0]

plt.plot(xa,ya,"-o",label="n=1")
plt.plot(xb,yb,"-o",label="n=3")
plt.plot(xc,yc,"-o",label="n=5")
plt.plot(xd,yd,"-o",label="n=10")
plt.legend()
plt.xlabel("successes")
plt.ylabel("probability")

plt.text(4.0, 0.45, "$\epsilon= 0.5$")
plt.savefig("binom_n.pdf")

```

```python
from scipy.stats import poisson
x = np.arange(0,20)

ya = poisson.pmf(x,1.0)
yb = poisson.pmf(x,3.0)
yc = poisson.pmf(x,5.0)
yd = poisson.pmf(x,10.0)

plt.plot(x,ya,"-o",label="$\lambda=1$")
plt.plot(x,yb,"-o",label="$\lambda=3$")
plt.plot(x,yc,"-o",label="$\lambda=5$")
plt.plot(x,yd,"-o",label="$\lambda=10$")

plt.legend()
plt.xlabel("count")
plt.ylabel("probability")
plt.savefig("poisson.pdf")



```

```python
from scipy.stats import norm

x = np.arange(0,20,0.1)
ya = norm.pdf(x,10.0,1.0)
yb = norm.pdf(x,10.0,2.0)
yc = norm.pdf(x,10.0,3.0)

plt.plot(x,ya,"-",label="$\sigma=1.0$")
plt.plot(x,yb,"-",label="$\sigma=2.0$")
plt.plot(x,yc,"-",label="$\sigma=3.0$")

plt.text(0.5, 0.37, "mean=10")
plt.xlabel("x")
plt.ylabel("probability density")
plt.legend()

plt.savefig("gaussian.pdf")


```

```python
from scipy import stats
plt.xlabel("x")
plt.ylabel("P(x)")
x = np.linspace(-3.0,3.0,100)
y = stats.norm.pdf(x,loc=0,scale=1)
plt.plot(x,y,"k-")
ya = stats.norm.pdf(-1.0,loc=0,scale=1)
yb = stats.norm.pdf(0.0,loc=0,scale=1)
plt.plot([-1,-1],[0,ya],"b--")
plt.plot([0,0],[0,yb],"k--")
plt.plot([1,1],[0,ya],"b--")
gca().set_xticks([-1,0,1])
gca().set_xticklabels(["$x_0 - \sigma$","$x_0$","$x_0 + \sigma$"])
gca().set_yticks([0,ya,yb])
gca().set_yticklabels(["0","$1/\sqrt{2\pi e}\sigma$","$1/\sqrt{2\pi}\sigma$"])
plt.savefig("stdunc.pdf", bbox_inches='tight')
```

```python
from scipy import stats
plt.xlabel("a")
plt.ylabel("$\Delta \chi^2$")
x = np.linspace(-3.0,3.0,100)
ypar = x**2
ychi = x**2 + 0.2*x**3 + 0.05*x**4
plt.plot(x,ychi,"k-", label="actual $\chi^2$")
plt.plot(x,ypar,"b--", label="parabola approximation")
gca().set_xticks([-1,0,1])
gca().set_xticklabels(["$a_0 - \sigma_a$","$a_0$","$a_0 + \sigma_a$"])
plt.ylim(0,10)
plt.legend()
plt.savefig("delchisq.pdf", bbox_inches='tight')
```

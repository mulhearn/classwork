```python
V = 12
R = 4

Vsrc = np.array([V,V])
Isrc = np.array([0,6])
Vld  = np.array([0,15])
Ild  = Vld / R 

plt.plot(Vld, Ild, "r-", label="load")
plt.plot(Vsrc, Isrc, "b-", label="ideal voltage source")
plt.plot(V,V/R,"ko")
plt.xlabel("V (V)")
plt.ylabel("I (mA)")
plt.legend()
plt.text(V-0.7,V/R+0.2,"Q")
plt.savefig("thev_ideal.pdf")

```

```python
R = 4
Vth = 12
Rth = 3
Vsrc = np.array([0,15])
Isrc = (Vth - Vsrc)/ Rth
Vld  = np.array([0,15])
Ild  = Vld / R 

Ix = Vth / (Rth + R)
Vx = R*Ix

plt.plot(Vth,0,"ko")
plt.plot(0,Vth/Rth,"ko")

plt.plot(0,Vth/Rth,"ko")
plt.plot(Vx,Ix,"ko")

plt.plot(Vld, Ild, "r-", label="load")
plt.plot(Vsrc, Isrc, "b-", label="non-ideal voltage source")
plt.xlabel("V (V)")
plt.ylabel("I (mA)")
plt.ylim(bottom=0)
plt.legend()
plt.text(0.5,4,"Isc")
plt.text(12,0.2,"Voc")
plt.text(Vx,Ix+0.3,"Q")
plt.savefig("thev_nonideal.pdf")

```

```python
Vth = 10
Rth = 5
Vsrc = np.array([0,12])
Isrc = (Vth - Vsrc)/ Rth

plt.plot(Vth,0,"ko")
plt.plot(0,Vth/Rth,"ko")

plt.plot(Vsrc, Isrc, "b-", label="non-ideal voltage source")
plt.xlabel("V (V)")
plt.ylabel("I (mA)")
plt.ylim(bottom=0)
plt.legend()
plt.text(0.5,2,"Isc")
plt.text(10,0.1,"Vth")
plt.savefig("thevenin.pdf")
```

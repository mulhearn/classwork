```python

def arrowed_spines(fig, ax):

    xmin, xmax = ax.get_xlim() 
    ymin, ymax = ax.get_ylim()

    # removing the default axis on all sides:
    for side in ['bottom','right','top','left']:
        ax.spines[side].set_visible(False)

    # removing the axis ticks
    plt.xticks([]) # labels 
    plt.yticks([])

    # get width and height of axes object to compute 
    # matching arrowhead length and width
    dps = fig.dpi_scale_trans.inverted()
    bbox = ax.get_window_extent().transformed(dps)
    width, height = bbox.width, bbox.height

    # manual arrowhead width and length
    hw = 1./40.*(ymax-ymin) 
    hl = 1./40.*(xmax-xmin)
    lw = 0.5 # axis line width
    ohg = 0.15 # arrow overhang

    # compute matching arrowhead length and width
    yhw = hw/(ymax-ymin)*(xmax-xmin)* height/width 
    yhl = hl/(xmax-xmin)*(ymax-ymin)* width/height

    # draw x and y axis
    ax.arrow(xmin, 0, xmax-xmin, 0., fc='k', ec='k', lw = lw, 
             head_width=hw, head_length=hl, overhang = ohg, 
             length_includes_head= True, clip_on = False) 

    ax.arrow(0, ymin, 0., ymax-ymin, fc='k', ec='k', lw = lw, 
             head_width=yhw, head_length=yhl, overhang = ohg, 
             length_includes_head= True, clip_on = False)

rz = 0.7
iz = 0.5

plt.figure(figsize=(4, 4), dpi=80)


plt.plot([rz,rz],[0,iz],"k--")
plt.plot([0,rz],[iz,iz],"k--")

plt.arrow(0,0,rz,iz,overhang=0.01,head_width=0.03,color="black",length_includes_head=True)







plt.xlim(0,1.0)
plt.ylim(0,1.0)
plt.text(0.95,0.03,"Re")
plt.text(0.02,0.95,"Im")

plt.text(rz-0.05,-0.05,"$r \, \cos \phi $")
plt.text(-0.15,iz,"$r \, \sin \phi $")

plt.text(0.075,0.02,"$\phi$")

plt.text(rz+0.01,iz+0.01,"$r \, e^{j \phi}$")






fig = plt.gcf()
fig.set_facecolor('white') 
ax = plt.gca()



arrowed_spines(fig, ax)

plt.savefig("complex.pdf")
```

```python
def diode(x):
    IS = 1.0
    VT = 0.025*10
    VL = 3*VT  
    ML = (IS/VT)*exp(VL/VT)
    IL = IS*(np.exp(VL/VT) - 1)    
    y=np.zeros(x.size)
    left = (x < VL)
    right  = (x >= VL)
    y[left] = IS*(np.exp(x[left]/VT) - 1)
    y[right] = IL + ML*(x[right]-VL)
    return y

plt.figure(figsize=(4, 4), dpi=80)
x = linspace(-2,1,100)
y = diode(x)
plt.plot(x,y)
plt.text(-2.5,3.2,"Current",rotation=90)
plt.xlabel("Voltage")
```

```python
def diode(x):
    N  = 2.0   # ideality factor
    IS = 5E-6  # 1N914 -> 5 uA
    VT = 0.025*N
    VB = -2.0  # fictitious
    y = IS*(np.exp(x/VT) - 1 - np.exp(-(x-VB)/VT))
    return y

plt.figure(figsize=(4, 4), dpi=80)
x = linspace(-3,1,100)
y = diode(x)
plt.plot(x,y)
plt.ylim(-0.2,1.0)
plt.xlim(-3,2)
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.plot([0,0],[-1.0,1.0],"k--")
#plt.plot([-10.0,10.0],[0,0],"k--")

gca().set_yticks([0])
gca().set_yticklabels(["0"])
gca().set_xticks([-2.5,0,0.6])
gca().set_xticklabels(["$V_{\\rm B}$","0","$V_{\\rm D}$"])


plt.savefig("breakdown.pdf")
```

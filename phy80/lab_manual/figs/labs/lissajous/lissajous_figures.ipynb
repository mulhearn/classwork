```python
t = np.linspace(0,1,num=1000)
w = 2*np.pi

A = 1.5

# Starting Pattern:
wx = 2*w
wy = 3*w
delta = 0

delta = np.pi * delta / 180
x = A*np.sin(wx*t+delta)
y = A*np.sin(wy*t)
plt.plot(x,y)
plt.axis('equal')
plt.xlabel("Vx (V)")
plt.ylabel("Vy (V)")
plt.savefig("pythonlissajous.pdf")

#plt.plot(t,x)
#plt.plot(t,y)
```
